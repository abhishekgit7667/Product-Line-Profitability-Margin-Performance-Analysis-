```python
#standarlized column names 
df.columns = (df.columns
.str.strip()
.str.lower()
.str.replace("/","_", regex = False )
.str.replace (" ", "_", regex = False )
.str.replace ("-","_", regex = False))

product_factory.columns = (product_factory.columns.
                           str.strip()
                           )
factory_coords.columns = (factory_coords.columns .str . lower ())
```
