```python
#convert numeric columns
key_cols = ["sales", "units", "gross_profit", "cost", "division", "product_name"]
df[key_cols].isnull().sum()
```

-remove rows where sales is missing or <= 0
-remove rows where cost is missing or < 0
-if gross_profit is missing and both sales/cost exist, calculate it
-if units is missing or <= 0, remove row for now because profit per unit will be needed
-remove rows where product_name or division is missing

```python
# Fill gross_profit if missing and sales/cost available
"""Agar gross_profit missing hai (NaN hai) aur sales aur cost available hai,
to gross_profit = sales - cost kar do,
warna jo pehle value hai wahi rehne do."""

df["gross_profit"] = np.where(
    df["gross_profit"].isna() & df["sales"].notna() & df["cost"].notna(),
    df["sales"] - df["cost"],
    df["gross_profit"]
)

# Remove rows with critical missing text fields if there is any missing value in the subset division and product name 
df = df.dropna(subset=["division", "product_name"])

# Remove invalid business rows if sales , cost , units grater than 0 keep otherwise delete 
df = df[df["sales"] > 0]
df = df[df["cost"] >= 0]
df = df[df["units"] > 0]
df = df[df["gross_profit"].notna()]
```

```python
#remove duplicates 
print("Duplicate rows before removal:", df.duplicated().sum())
#how much gross profit filled by gross profit = sales - cost
print ("filling filled :", df["gross_profit"].isna().sum())
```

gross_profit ≈ sales - cost

```python
#we check profit validation
df["calculated_profit"] = df["sales"] - df["cost"]
df["profit_diff"] = df["gross_profit"] - df["calculated_profit"]
df["profit_diff_abs"] = df["profit_diff"].abs()
df["profit_diff_abs"] = df["profit_diff_abs"].round(2)
```

```python
tolerance = 0.01

invalid_profit_rows = df[df["profit_diff_abs"] > tolerance]
print("Rows with profit mismatch:", invalid_profit_rows.shape[0])
```

Merge product-factory mapping

```python
product_factory["product_name"] = product_factory["product_name"].astype(str).str.strip()
product_factory["division"] = product_factory["division"].astype(str).str.strip()

factory_coords["factory"] = factory_coords["factory"].astype(str).str.strip()

df["product_name"] = df["product_name"].astype(str).str.strip()
df["division"] = df["division"].astype(str).str.strip()
```

```python
df = df.merge(
    product_factory,
    on=["division", "product_name"],
    how="left"
)
```
