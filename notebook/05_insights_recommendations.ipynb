```python
import pandas as pd

product_summary = pd.read_csv("D:/nassau_profitability_project/data/processed/product_summary.csv")
division_summary = pd.read_csv("D:/nassau_profitability_project/data/processed/division_summary.csv")
monthly_summary = pd.read_csv("D:/nassau_profitability_project/data/processed/monthly_summary.csv")
factory_summary = pd.read_csv("D:/nassau_profitability_project/data/processed/factory_summary.csv")
pareto_profit = pd.read_csv("D:/nassau_profitability_project/outputs/tables/pareto_profit_table.csv")
pareto_revenue = pd.read_csv("D:/nassau_profitability_project/outputs/tables/pareto_revenue_table.csv")
```

2. Repricing candidates

These are products with:

-high sales
-low margin

This means demand is there, but margin quality is weak.

```python
sales_threshold = product_summary["sales"].median()
margin_threshold = product_summary["gross_margin_pct"].median()

repricing_candidates = product_summary[
    (product_summary["sales"] >= sales_threshold) &
    (product_summary["gross_margin_pct"] < margin_threshold)
].sort_values(["sales", "gross_margin_pct"], ascending=[False, True])

repricing_candidates.head(10)
```

```python
repricing_candidates.to_csv("D:/nassau_profitability_project/outputs/tables/repricing_candidates.csv", index=False)
```

3.  Promotion candidates

These are products with:

-high margin
-low sales

This means product quality is strong, but market performance is underutilized.

```python
promotion_candidates = product_summary[
    (product_summary["gross_margin_pct"] >= margin_threshold) &
    (product_summary["sales"] < sales_threshold)
].sort_values("gross_margin_pct", ascending=False)

promotion_candidates.head(10)
```

```python
promotion_candidates.to_csv("D:/nassau_profitability_project/outputs/tables/promotion_candidates.csv", index=False)
```

4. Discontinuation review candidates

These are products with:

low sales
low profit
weak margin

These products may be hurting portfolio efficiency.

```python
profit_threshold = product_summary["gross_profit"].median()

discontinue_candidates = product_summary[
    (product_summary["sales"] < sales_threshold) &
    (product_summary["gross_profit"] < profit_threshold) &
    (product_summary["gross_margin_pct"] < margin_threshold)
].sort_values(["gross_profit", "sales"], ascending=True)

discontinue_candidates.head(10)
```

```python
discontinue_candidates.to_csv("D:/nassau_profitability_project/outputs/tables/discontinue_review_candidates.csv", index=False)
```

5) Cost renegotiation candidates

We did not save cost_ratio in product summary earlier, so now create it first.

```python
product_summary["cost_ratio"] = product_summary["cost"] / product_summary["sales"]
```

Now identify products where:

-cost ratio is high
-margin is low

```python
cost_ratio_threshold = product_summary["cost_ratio"].median()

cost_renegotiation_candidates = product_summary[
    (product_summary["cost_ratio"] > cost_ratio_threshold) &
    (product_summary["gross_margin_pct"] < margin_threshold)
].sort_values(["cost_ratio", "gross_margin_pct"], ascending=[False, True])

cost_renegotiation_candidates.head(10)
```

```python
cost_renegotiation_candidates.to_csv("D:/nassau_profitability_project/outputs/tables/cost_renegotiation_candidates.csv", index=False)
```

6) Star products

These are your strongest products.

```python
star_products = product_summary[
    (product_summary["gross_profit"] >= profit_threshold) &
    (product_summary["gross_margin_pct"] >= margin_threshold)
].sort_values(["gross_profit", "gross_margin_pct"], ascending=False)

star_products.head(10)
```

```python
star_products.to_csv("D:/nassau_profitability_project/outputs/tables/star_products.csv", index=False)
```

7) Revenue vs profit imbalance insight table

Create a difference column.

```python
division_summary["contribution_gap"] = (
    division_summary["profit_contribution_pct"] - division_summary["revenue_contribution_pct"]
)

division_imbalance = division_summary.sort_values("contribution_gap")
division_imbalance
```

Interpretation:

negative gap → revenue is strong but profit is weaker
positive gap → division is more efficient than its revenue share suggests

```python
division_imbalance.to_csv("D:/nassau_profitability_project/outputs/tables/division_imbalance.csv", index=False)
```

8) Profit concentration risk insight

Find how many products contribute 80% of profit.

```python
profit_80_count = (pareto_profit["cumulative_profit_pct"] <= 80).sum()
revenue_80_count = (pareto_revenue["cumulative_sales_pct"] <= 80).sum()

total_products = product_summary.shape[0]

profit_80_pct_products = (profit_80_count / total_products) * 100
revenue_80_pct_products = (revenue_80_count / total_products) * 100

print("Products contributing 80% of profit:", profit_80_count)
print("Products contributing 80% of revenue:", revenue_80_count)
print("Percent of products contributing 80% of profit:", round(profit_80_pct_products, 2))
print("Percent of products contributing 80% of revenue:", round(revenue_80_pct_products, 2))
```

```python
# store in small table 
concentration_summary = pd.DataFrame({
    "metric": ["80% Profit Concentration", "80% Revenue Concentration"],
    "product_count": [profit_80_count, revenue_80_count],
    "percent_of_total_products": [profit_80_pct_products, revenue_80_pct_products]
})

concentration_summary
```

```python
concentration_summary.to_csv("D:/nassau_profitability_project/outputs/tables/concentration_summary.csv", index=False)
```

9) Factory efficiency insight

Add contribution metrics to factory summary.

```python
factory_summary["revenue_contribution_pct"] = (
    factory_summary["sales"] / factory_summary["sales"].sum()
) * 100

factory_summary["profit_contribution_pct"] = (
    factory_summary["gross_profit"] / factory_summary["gross_profit"].sum()
) * 100

factory_summary["gross_margin_pct"] = (factory_summary["gross_profit"] / factory_summary["sales"]) * 100

factory_summary.sort_values("gross_profit", ascending=False)
```

```python
factory_summary.to_csv("D:/nassau_profitability_project/outputs/tables/factory_efficiency_summary.csv", index=False)
```

10) Create action recommendation column

This is very useful for dashboard and report.

```python
def recommendation(row):
    if row["sales"] >= sales_threshold and row["gross_margin_pct"] < margin_threshold:
        return "Reprice"
    elif row["cost_ratio"] > cost_ratio_threshold and row["gross_margin_pct"] < margin_threshold:
        return "Renegotiate Cost"
    elif row["gross_margin_pct"] >= margin_threshold and row["sales"] < sales_threshold:
        return "Promote"
    elif row["sales"] < sales_threshold and row["gross_profit"] < profit_threshold:
        return "Discontinue Review"
    else:
        return "Monitor"

product_summary["recommended_action"] = product_summary.apply(recommendation, axis=1)
product_summary[["product_name", "division", "gross_profit", "gross_margin_pct", "recommended_action"]].head(10)
```

```python
product_summary.to_csv("D:/nassau_profitability_project/outputs/tables/product_summary_with_actions.csv", index=False)
```
